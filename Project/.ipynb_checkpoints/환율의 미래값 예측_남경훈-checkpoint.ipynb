```python
#열이름 수정
usa = usa.rename(columns={'일자':'date',
                        '송금':'send',
                        'Unnamed: 2':'receipt',
                        '현찰':'buy',
                        'Unnamed: 4':'sale',
                        'T/C사실때':'T/C',
                        '매매기준율':'sbr',
                        '한국은행기준율':'kbbr',
                        '대미환산율':'usr',
                        '전일대비':'pde'
                   })
usa.head(10)
```

## 특정 문자열 제거 

```python
for i in usa.columns:
    if i != 'usr':
        usa[i] = usa[i].str.replace(",","")
        usa[i] = usa[i].str.replace("/td>","")
        usa[i] = usa[i].str.replace("하락","")
        usa[i] = usa[i].str.replace("상승","")
        
usa['date'] = usa['date'].str.replace(".","")
display(usa)
```

```python
import warnings
warnings.filterwarnings("ignore")
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

## 기존 전처리 데이터 사용.
## datetime 으로 변환.

```python
#모듈 설치 
from fbprophet import Prophet
#from datetime import datetime

usa['date'] = pd.to_datetime(usa['date'])
print(usa['date'].head(10))
```

## 데이터 확인 및 처리

```python
#데이터 타입 확인
print(usa.dtypes)
#타입 변경
print(usa.dtypes)
print(usa.head(10))
```

## 데이터 전처리

```python
df = pd.DataFrame({'ds':usa.date, 'y':usa['sbr']})
df.reset_index(inplace = True)
del df['index']
df.head()
```

## 모델적용

```python
m = Prophet()
m.fit(df);
```

## 예측할 데이터프레임 만들기

```python
#예측할 데이터 프레임 만들기.
future = m.make_future_dataframe(periods=365)
future.tail()
```

## 예측할 데이터프레임 예측

```python
forecast = m.predict(future)
```

## 예측한 데이터 확인

```python
#Prophet 으로 실제 예측 및 결과값 확인
#yhat = 예측 결과 , yhat_lower = 예측 최소값, yhat_upper = 예측 최대값
forecast[['ds','yhat','yhat_lower','yhat_upper']].head(10)
```

## 예측한 결과값의 plot

```python
#검은 반점이 원 주가, 파란 실선이 예측값, 엷은 파란색의 범위는 유효범위
m.plot(forecast);
```

```python
#trend 분석 시각화
m.plot_components(forecast);
```

## 상한가를 주고 나타낸 plot

```python
future = m.make_future_dataframe(periods = 365)

future['cap'] = 1400
future['floor'] = 1000

forecast = m.predict(future)
from fbprophet.plot import add_changepoints_to_plot

fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), m, forecast)
```
