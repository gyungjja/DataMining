```python
import warnings
warnings.filterwarnings("ignore")
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
usa = pd.read_csv("usa.csv")
usa.head()
```

```python
#NaN 처리
usa = usa.dropna(how='any')
display(usa)
```

```python
#열이름 수정
usa = usa.rename(columns={'일자':'date',
                        '송금':'send',
                        'Unnamed: 2':'receipt',
                        '현찰':'buy',
                        'Unnamed: 4':'sale',
                        'T/C사실때':'T/C',
                        '매매기준율':'sbr',
                        '한국은행기준율':'kbbr',
                        '대미환산율':'usr',
                        '전일대비':'pde'
                   })
display(usa)
```

```python
#특정 문자열 제거
for i in usa.columns:
    if i != 'usr':
        usa[i] = usa[i].str.replace(",","")
        usa[i] = usa[i].str.replace("/td>","")
        usa[i] = usa[i].str.replace("하락","")
        usa[i] = usa[i].str.replace("상승","")
        
usa['date'] = usa['date'].str.replace(".","")
display(usa)
```

```python
#모듈 설치 
from fbprophet import Prophet
#from datetime import datetime

usa['date'] = pd.to_datetime(usa['date'])
print(usa['date'])
```

```python
#데이터 타입 확인
print(usa.dtypes)
#타입 변경
usa.iloc[:,1:11].astype(float)
print(usa.dtypes)
print(usa)
```

```python
df = pd.DataFrame({'ds':usa.date, 'y':usa['sbr']})
df.reset_index(inplace = True)
del df['index']
df.head()
```

```python
m = Prophet()
m.fit(df);
```

```python
#예측할 데이터 프레임 만들기.
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast.dtypes
```

```python
#Prophet 으로 실제 예측 및 결과값 확인
#yhat = 예측 결과 , yhat_lower = 예측 최소값, yhat_upper = 예측 최대값
forecast[['ds','yhat','yhat_lower','yhat_upper']]
```

```python
display(forecast)
```

```python
kbbr = usa.kbbr.astype('float')
kbbr
sbr = usa.sbr.astype('float')
sbr
#날짜별 kbbr 그래프
LINE = plt.plot(usa.date,kbbr.tolist())
plt.setp(LINE,linewidth=3.0)
plt.ylim(1000,1300)
plt.show()
```

```python
#검은 반점이 원 주가, 파란 실선이 예측값, 엷은 파란색의 범위는 유효범위
m.plot(forecast);
```

```python
#trend 분석 시각화
m.plot_components(forecast);
```
