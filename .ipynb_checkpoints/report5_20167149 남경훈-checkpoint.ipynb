# [실습 #5]merging, grouping, sampling 예제 확인

```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
#merging
d1 = pd.DataFrame({'key': ['K0','K1','K2','K3'],
                  'A': ['A0', 'A1', 'A2', 'A3'],
                  'B': ['B0', 'B1', 'B2', 'B3']})
d2 = pd.DataFrame({'key': ['K0','K1','K2','K3'],
                  'C': ['C0', 'C1', 'C2', 'C3'],
                  'D': ['D0', 'D1', 'D2', 'D3']})
display(d1)
display(d2)

d3 = d1.append(d2)
display(d3)
d4 = d1.append(d2, ignore_index = True)
display(d4)

d5 = pd.merge(d1, d2, on=['key'])
display(d5)
```

```python
#Grouping
df = pd.DataFrame({'name': ['kim','min','jin','song','lee'],
                      'age': [20, 30, 25, 40, 23],
                      'sex': ['F', 'M', 'F', 'M', 'M'],
                      'score': [100, 50, 80, 90, 98],
                      'class': pd.Categorical([1, 2, 2, 3, 1 ,])})
display(df)

print("## df.sum() ##")
print(df.sum())

print("## group (sex) ##")
grouped = df.groupby('sex')
for name, group in grouped:
    print(name)
    print(group)
    
#그룹 추출
f = grouped.get_group(name='F')
m = grouped.get_group(name='M')
print("## group (female) ##")
display(f)
print("## group (male) ##")
display(m)

#그룹별 연산
print('#grouped.sum()=')
print(grouped.sum())
print('#grouped.mean()=')
print(grouped.mean())

#그룹에 대한 요약 통계
r1 = grouped.agg(['sum', 'mean', 'std'])
print('#grouped.agg()=\n', r1)
```

```python
#Sample
df = pd.DataFrame({ 'name' : ['kim','min','jin','song','lee'],
                    'age' : [20,30,25,40,23],
                    'sex' : ['F','M','F','M','M'],
                    'score' : [100,50,80,90,98],
                    'class' : pd.Categorical([1,2,2,3,1,])})

dfs1 = df.sample(n=3)
dfs2 = df.sample(n=3, random_state = 1)
dfs3 = df.sample(n=3,weights='class')

display(dfs1)
display(dfs2)
display(dfs3)
```

# 2)iris 데이터셋을 이용하여 다음을 처리
# iris 데이터 읽어오기

```python
iris = pd.read_csv("irisData.csv")
iris
```

## (1) Species을 기준으로 groupby 처리

```python
print("## group (Species) ##")
grouped = iris.groupby('Species')
for name, group in grouped:
    print(name)
    print(group)
```

## (2)(1)의 결과를 이용하여 Species별 평균, 표준편차 확인

```python
print('#grouped.mean() = ')
print(grouped.mean())
print('#grouped.std() = ')
print(grouped.std())

#그룹에 대한 평균,표준편차
r1 = grouped.agg(['mean','std'])
print('#grouped.agg() = \n', r1)
```

## (3)(1)의 결과를 이용하여 Species(종별) 로 분리

```python
s = grouped.get_group(name="setosa")
ve = grouped.get_group(name = "versicolor")
vi = grouped.get_group(name = "virginica")
print("##group (setosa) ##")
display(s)
print("##group (versicolor) ##")
display(ve)
print("##group (virginica) ##")
display(vi)
```

## (4)분리된 3개의 종에서 각각 20개씩 랜덤 추출하여 결합

```python
dfs1 = s.sample(n=20)
dfs2 = ve.sample(n=20, random_state = 1)
dfs3 = vi.sample(n=20, random_state = 10)

print("## Sample (setosa) ##")
display(dfs1)
print("## Sample (versicolor) ##")
display(dfs2)
print("## Sample (virginica) ##")
display(dfs3)

iris20 = dfs1.append(dfs2)
iris20 = iris20.append(dfs3)
print("## Sample Append ##")
display(iris20)
```

# - 실습 #5의 iris 데이터로부터 추출된 결과를 이용하여 아래문제를 해결하세요.

## 실습 #5 : iris 데이터를 종별로 분리하여 20개씩 샘플링한 결과를 결합

### (1)3개 종에 대한 boxplot(subplot 이용)

```python
import matplotlib.pyplot as plt
print("First boxplot = setosa\n")
plt.subplot(3,1,1)
dfs1.boxplot()
print("Second boxplot = versicolor\n")
plt.subplot(3,1,2)
dfs2.boxplot()
print("Third boxplot = virginica\n")
plt.subplot(3,1,3)
dfs3.boxplot()
```

### (2)3개 종에 대한 histogram

```python
dfs1.plot.hist()
plt.title("setosa histogram")
dfs2.plot.hist()
plt.title("versicolor histogram")
dfs3.plot.hist()
plt.title("virginica histogram")
```

### (3)3개 종에 대한 scatter plot을 결합

#### (3-1)Sepal width, Petal width 이용

```python
dfs1.plot.scatter(x='Sepal width', y='Petal width', c='Blue')
plt.title("setosa scatter plot")
dfs2.plot.scatter(x='Sepal width', y='Petal width', marker='^', s=60)
plt.title("versicolor scatter plot")
dfs3.plot.scatter(x='Sepal width', y='Petal width', c='Red')
plt.title("virginica scatter plot")
```

#### (3-2)Sepal length,Petal length 이용

```python
dfs1.plot.scatter(x='Sepal length', y='Petal length', c='Blue')
plt.title("setosa scatter plot")
dfs2.plot.scatter(x='Sepal length', y='Petal length', marker='^', s=60)
plt.title("versicolor scatter plot")
dfs3.plot.scatter(x='Sepal length', y='Petal length', c='Red')
plt.title("virginica scatter plot")
```

### (4)결합된 데이터의 scatter matrix

```python
pd.plotting.scatter_matrix(iris20, s=60, diagonal='hist', c=iris20['Sepal length'], cmap='viridis')
pd.plotting.scatter_matrix(iris20, s=60, diagonal='hist', c=iris20['Sepal width'], cmap='viridis')
pd.plotting.scatter_matrix(iris20, s=60, diagonal='hist', c=iris20['Petal length'], cmap='viridis')
pd.plotting.scatter_matrix(iris20, s=60, diagonal='hist', c=iris20['Petal width'], cmap='viridis')
```

### (5) 결합된 데이터의 평균에 대한 bar,pie plot

```python
#결합된 데이터의 평균 구하기
iris20_mean = grouped.agg(['mean'])
print('#grouped.mean() =\n',iris20_mean)

#barplot
iris20_mean.plot.bar(title="Bar plot",legend = True)

#pie
iris20_mean.plot.pie(y = 'Sepal length',title="Pie plot")
iris20_mean.plot.pie(y = 'Sepal width',title="Pie plot")
iris20_mean.plot.pie(y = 'Petal length',title="Pie plot")
iris20_mean.plot.pie(y = 'Petal width',title="Pie plot")


```
