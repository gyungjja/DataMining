```python
import pandas as pd
from IPython.display import display

iris = pd.read_csv('irisData.csv')
display(iris)
```

```python
print(iris.index)
print(iris.columns)
iris.describe()

iris["Species"]

#print((iris["Petal length"]), ascending = False)
iris["Petal length"][::-1]

```

```python
iris[iris.Species == 'setosa']
iris[iris.iloc[:,0] <= 5]
iris[iris.iloc[:,4] == 'setosa']
```

```python
iris1 = iris.drop(['Petal length'],axis = 1)
iris1.iloc[:,1:]
iris2 = iris.drop(['Sepal width'], axis = 1)
iris2.iloc[:,0:2]
```

```python
ratio = iris['Sepal length'] / iris['Sepal width']
ratio
ratio2 = iris['Petal length'] / iris['Petal width']
ratio2

iris['ratio'] = ratio
iris
iris['ratio2'] = ratio2
iris
```

```python
grouped = iris.groupby('Species')
for name,group in grouped:
    display(name)
    display(group)
```

```python
r1 = grouped.agg(['mean','std'])
r1
```

```python
se = grouped.get_group(name = 'setosa')
se
ve = grouped.get_group(name = 'versicolor')
vi = grouped.get_group(name = 'virginica')
```
