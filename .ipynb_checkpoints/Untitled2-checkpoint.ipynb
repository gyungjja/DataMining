```python
import numpy as np
import pandas as pd
from IPython.display import display

#Series, numpy array로 생성
s1 = pd.Series(np.random.randn(5))
s2 = pd.Series(np.random.randn(5), index=['a','b','c','d','e'])

#Seres, 파이썬 dicts, KEY-VALUE로 생성
d = {'b':1,'a':0,'c':2}
s3 = pd.Series(d)

s4 = pd.Series(5., index=['a', 'b', 'c', 'd', 'e'])

print("s1 =",s1)
print("s2=", s2)
print("s3=", s3)
print("s4=", s4)

print("s1.ndim=", s1.ndim)
print("s1.shape=", s1.shape)
print("s1.shape=", s1.dtype)
```

```python
print("s1[0]=", s1[0])
print("s1[:2]=", s1[:2])
print("s1[2:4:2]=", s1[2:4:2])
print("s1[3:]=", s1[3:])
print("s1[s1 > s1.median()]=", s1[s1 > s1.median()])
print("s3 + 2 =", s3 + 2)
print("s3 + s4 =", s3 + s4)
print("s3.sum() =", s3.sum())

print("s2['a']=", s2['a'])
print("s2.get('a')=", s2.get('a'))
print("'d' in s2 =",'d' in s2)
```

```python
data = np.floor(np.random.randn(7, 3) * 10)
idx = np.arange(7) + 1
col = list('ABC')
df1 = pd.DataFrame(data, index = idx, columns=col)
display(df1)

dates = pd.date_range('20190401', periods=7)
df2 = pd.DataFrame(data)
df2.columns = col
display(df2)

d = {'one': [1., 2., 3., 4.],
     'two': [4., 3., 2., 1.]}
df3 = pd.DataFrame(d)
df4 = pd.DataFrame(d, index=['a', 'b', 'c', 'd'])
display(df3)
display(df4)
```

```python
#(1)DataFrame 생성
d2 = {'one' : pd.Series([1., 2., 3.], index=['a','b','c']),
      'two' : pd.Series([1., 2., 3., 4.], index=['a','b','c','d'])}
df5 = pd.DataFrame(d2)
print("df5.dtypes = ", df5.dtypes)
display(df5)

data = np.zeros((2, ), dtype = [('A', 'i4'), ('B', 'f4'), ('C', 'a10')])
data[:] = [(1, 2., 'Hello'), (2, 3., "world")]
df6 = pd.DataFrame(data)
display(df6)

df7 = pd.DataFrame({'A' : 1.,
                    'B': pd.Timestamp('20190401'),
                    'C': pd.Series(1, index=list(range(4)), dtype='float32'),
                    'D': np.array([3] * 4, dtype='int32'),
                    'E': pd.Categorical(["test","train","test","train"]),
                    'F' : 'foo'})

display(df7)
```

```python
#2)DataFrame 확인(Viewing)
data = np.floor(np.random.randn(8,3) * 10)
col = list('XYZ')
dates = pd.date_range('20190301', periods=8)
df = pd.DataFrame(data,index = dates, columns=col)
display(df)

print("df=", df)
print("df.head()=", df.head())
print("df.tail(3)=", df.tail(3))

print("df.dtypes = ", df.dtypes)
print("df.index = ", df.index)
print("df.columns = ", df.columns)
```

```python
#통계요약
df.describe()

#Sorting by an axis:
dfsi = df.sort_index(axis=0, ascending=False)
display(dfsi)
dfvy = df.sort_values(by='Y')
display(dfvy)
```

```python
#(3) DataFrame,Indexing/Selection
d = {'one':pd.Series([1.,2.,3.], index=['a','b','c']),
     'two':pd.Series([1.,2.,3.,4.], index=['a','b','c','d'])}

df1 = pd.DataFrame(d)
display(df1)

print("df1[0:2]=", df1[0:2])
print("df1['a':'c'] = ", df1['a':'c'])

print("df1.loc['a'] =",df1.loc['a'])
print("df1.loc['b':]=",df1.loc['b':])
print("df1.loc[['a','c']]=",df1.loc[['a','c']])
print("df1.iloc[1]=",df1.iloc[1])
print("df1.iloc[1:3]=",df1.iloc[1:3])

print("df1.iloc[:, 1]=", df1.iloc[:, 1])
print("df1.iloc[:, 1:]=", df1.iloc[:, 1:])

print("df1['one']=", df1['one'])
print("df1['two']['a'] = ", df1['two']['a'])
print("df1['two'][2:] = ",df1['two'][2:])

print("df1[df1.one >=2 ]=",df1[df1.one>=2])
print("pd.isna(df1.one) = ", pd.isna(df1.one))

print("df1.one.isin([1,3] = ",df1.one.isin([1,3]))
print("df1.isin([2,3]=",df1.isin([2,3]))
mask = df1.two.isin([1,3])
print("df1.loc[df1.two.isin([1,3]])=",df1.loc[mask])
```
