# 1)Heart.csv 파일 열기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

heart = pd.read_csv('Heart.csv')
display(heart)
```

# 2)데이터 타입들, 상하위 5개 데이터 확인

```python
print("데이터 타입 = \n",heart.dtypes)
print("상위 데이터 5개 확인 = \n",heart.head(5))
print("하위 데이터 5개 확인 = \n",heart.tail(5))
```

# 3)Unnamed: 0 열 삭제(원본에서 삭제)

```python
heart = heart.drop(['Unnamed: 0'], axis = 1)
heart
```

# 4) AHD 변경, No = 0 , Yes = 1

```python
heart.loc[heart.AHD == 'No','AHD'] = 0
heart.loc[heart.AHD == 'Yes','AHD'] = 1

display(heart)
```

# 5) 통계요약

```python
heart.describe()

count = heart.count()
csum = heart.sum()
rsum = heart.sum(axis = 1)
mean = heart.mean()
min1 = heart.min()
max1 = heart.max()

heart2 = heart.copy()
heart2.loc['count'] = count
heart2.loc['csum'] = csum
heart2.loc['mean'] = mean
heart2.loc['min'] = min1
heart2.loc['max'] = max1
heart2['rsum'] = rsum
display(heart2)
```

# 6) 누락된 값 있는 행 삭제

```python
heart.dropna(how='any')
```

```python
print("=========나이와 최대심박수 상관계수========\n")
ax = heart.plot.scatter(x='Age',y='MaxHR',c='Blue' , label = "Age&MaxHR")
print("=========나이와 혈압 상관계수========\n")
ax = heart.plot.scatter(x='Age',y='RestBP', c='Red' , label = "Age&RestBP", ax=ax)
```

# 11)속성중 4개만 선택하여 심장병과의 관계 분석 scaater_matrix()

```python
#RestBP, MaxHR, ExAng, Thal

heart3 = heart[['RestBP','MaxHR','Oldpeak','Chol']]
heart3_AHD = heart['AHD']
#heart3_AHD
pd.plotting.scatter_matrix(heart3, s=60, diagonal='hist', c=heart['AHD'] , cmap='viridis')
```

# 12) 4개 특징의 분포 확인(boxplot)

```python
heart4 = heart[['RestBP','MaxHR','Oldpeak','Chol']]
heart4.boxplot()
```

# 13) target(AHD)을 기준(groupby)으로 최대,최소,평균, 표준편차 계산

```python
grouped = heart.groupby('AHD')
print("target을 기준으로 최대,최소,평균,표준편차")
display(grouped.max())
display(grouped.min())
display(grouped.mean())
display(grouped.std())
```

# 14) 4개 특징과 target(AHD)를 이용하여 학습데이터, 테스트데이터 집합 준비(테스트 데이터는 전체에 20%)

```python
from sklearn.model_selection import train_test_split
#테스트 데이터 20%
X_train, X_test, y_train, y_test = train_test_split(
    heart4, heart['AHD'],test_size = 0.2, random_state = 0)

print("X_train 크기: {}".format(X_train.shape))
print("y_train 크기: {}".format(y_train.shape))

print("X_test 크기: {}".format(X_test.shape))
print("y_test 크기: {}".format(y_test.shape))
```

# 15) 1-NN ~ 9-NN 테스트 성능 비교(그래프로 확인)

```python
from sklearn.neighbors import KNeighborsClassifier

#테스트 데이터 20%
#X_train, X_test, y_train, y_test = train_test_split(
#    heart4, heart['AHD'],test_size = 0.2, random_state = 0)


training_accuracy = []
test_accuracy = []
#1에서 9까지 n_neighbors를 적용.
neighbors_settings = range(1,10)

for n_neighbors in neighbors_settings:
    #모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train,y_train)
    #훈련세트 정확도 저장
    training_accuracy.append(clf.score(X_train,y_train))
    #일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_setting, training_accuracy, label = "train accuracy")
plt.plot(neighbors_setting, test_accuracy, label = "test accuracy")
plt.ylabel("accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```
