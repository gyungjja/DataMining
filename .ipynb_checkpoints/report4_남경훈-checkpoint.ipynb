## 2)데이터 설명
##    -데이터 집합 및 속성 설명

### Attribute(열 특성)

- age: 연령
- sex: 성별(1 = 남성, 0 = 여성)
- cp: 가슴 통증 유형(1: 일반적인 혈관 형성 2: 비정형 혈관 3: 비혈관 질환 4: 무증상 )
- Trestbps: 휴면 혈압(병원 입원 시 mm Hg)
- chol: 혈청 콜레스트롤(mg/dl)
- fbs: (혈당 > 120 mg/dl) (1 = 참, 0 = 거짓)
- resecg: 휴식 심전도의 결과(0: 정상
                        1: ST-T 파형 이상(T 파형 반전 및/또는 ST 상승 또는 하강이 0.05mV를 초과함)
                        2: 에스테스의 기준에 의한 발생가능하거나 분명한 좌심실비대증을 보여준다.

- talach: 최대 심장 박동수 달성
- exang: 운동 유도 혈관부(1 = 예, 0 = 아니오)
- oldpeak = 휴식과 관련된 운동에 의해 유발된 ST 우울증(1: 상향 조정 2: 플랫 3: 다운스트림)
- slope: 피크 연습 ST 세그먼트의 기울기
- ca: 푸로사피 색상의 주요 용기 수(0-3)
- thal: 3 = 정상, 6 = 고정 결함, 7 = 반전 가능한 결함
- target: 심장병 진단(양성 질환 상태 1,0으로 나누어짐 1=양성 0=음성)


### 3)데이터 분석 준비:노트북파일

### (1)dataframe에 파일 읽어오기

```python
df = pd.read_csv("heart.csv")
display(df)
```

### (2)행레이블,열레이블 확인

```python
print("df.index = ", df.index)
print("df.columns = ", df.columns)
```

### (3)상위,하위데이터 확인

```python
print("상위데이터 5개확인 = \n")
display(df.head())
print("하위데이터 5개확인 = \n")
display(df.tail(5))
```

### (4)통계요약확인

```python
print("통계요약 = \n",df.describe())

count = df.count()
csum = df.sum()
rsum = df.sum(axis=1)
mean = df.mean()
min1 = df.min()
max1 = df.max()

df2 = df.copy()
df2.loc['count'] = count
df2.loc['csum'] = csum
df2.loc['mean'] = mean
df2.loc['min'] = min1
df2.loc['max'] = max1
df2['rsum'] = rsum

display(df2)
```

### (5)Nan이 속한 데이터 삭제

```python
df.dropna(how='any') #누락된 값 있는 행 삭제
```

### (6)sex속성을 변경 0(female),1(male)

```python
df.loc[df.sex == 1,'sex'] = 'female'
df.loc[df.sex == 0,'sex'] = 'male'

display(df)
```

### (7)여성데이터와 남성데이터만 추출

```python
print("여성데이터만 추출 = \n")
display(df[df.sex == 'female'])

print("남성데이터만 추출 = \n")
display(df[df.sex == 'male'])
```

### (8)여성데이터의 심장병데이터만 추출

```python
#심장병데이터 target 추출
target = df[df.sex=='female']['target']
display(target)
```

### (9)40대의 심장병 추출

```python
print("40대 환자들 추출 = \n")
display(df[(df.age >= 40) & (df.age < 50)])

print("40대 환자들 심장병 추출 = \n")
display(df[(df.age >= 40) & (df.age < 50)]['target'])
```

### (10)40미만 , 40-55, 50초과 연령에 대한 개수, 평균 확인

```python
#40미만 연령 개수,평균 확인.
df3 = df.copy()
mask = (df.age < 40)
df3 = df3[mask]

df3.loc['count'] = df3.count()
df3.loc['mean'] = df3.mean()
print("40미만 연령 개수,평균 확인 = \n")
display(df3)
```

```python
#40 - 55 연령 개수,평균 확인.
df4 = df.copy()
mask = (df.age >= 40) & (df.age <= 55)
df4 = df4[mask]

df4.loc['count'] = df3.count()
df4.loc['mean'] = df3.mean()
print("40 ~ 55 연령 개수,평균 확인 = \n")
display(df4)
```

```python
#50초과 연령 개수,평균 확인.
df4 = df.copy()
mask = (df.age > 50)
df4 = df4[mask]

df4.loc['count'] = df3.count()
df4.loc['mean'] = df3.mean()
print("50초과 연령 개수,평균 확인 = \n")
display(df4)
```
