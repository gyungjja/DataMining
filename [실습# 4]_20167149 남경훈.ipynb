## 2.File IO 연습

## 1)irisData.csv 파일 읽기

```python
#csv 파일 읽기
iris = pd.read_csv('irisData.csv')

df = pd.DataFrame(iris)
display(df)
```

## 2)데이터 살펴보기

### (1)상위, 하위 5줄 확인

```python
print("상위 5줄 확인=\n",df.head())
print("하위 5줄 확인=\n",df.tail(5))
```

### (2)index, column 살펴보기

```python
print("index = \n", df.index) #행 레이블
print("column = \n", df.columns) #열 레이블
```

### (3) 통계값 확인

```python
df.describe()

dfsi = df.sort_index(axis=0, ascending=False)
display(dfsi)

dfvy = df.sort_values(by = 'Sepal length')
display(dfvy)

count = df.count()
csum = df.sum()
rsum = df.sum(axis=1)
mean = df.mean()
min1 = df.min()
max1 = df.max()

df2 = df.copy()
df2.loc['count'] = count
df2.loc['csum'] = csum
df2.loc['mean'] = mean
df2.loc['min'] = min1
df2.loc['max'] = max1
df2['rsum'] = rsum

df2['Species'] = df2['Species'][:-4] # 문자열로된 열 제거

display(df2)
```

### (4) iris의 Species 속성값확인

```python
print("Species 속성값 확인 = \n" , df.iloc[:, 4])
```

### (5)iris의 꽃잎 길이(Petal length)의 역순으로 확인

```python
print("꽃잎 길이(Petal length) = \n",df.iloc[:, 3])
#print("꽃잎 길이(Petal length) 역순으로 확인 = \n",df.sort_index(axis = 0, by='Petal length', ascending=False))
print("꽃잎 길이(Petal length) = \n",df['Petal length'][::-1])
```

### (4)데이터 추가

### (1)꽃받침 길이와 꽃받침 폭의 비율을 새로운 속성으로 추가(Sepal)

```python
df_ratio = df.copy()
df_ratio
df_ratio['Ratio'] = df['Sepal length'] / df['Sepal width']

print("꽃받침 길이와 꽃받침 폭의 비율 = \n",df_ratio)
```

### (2)꽃잎 길이와 꽃잎 폭의 비율을 새로운 속성으로 추가(Petal)

```python
df_ratio2 = df.copy()
df_ratio2
df_ratio2['Ratio'] = df['Petal length'] / df['Petal width']

print("꽃잎 길이와 꽃잎 폭의 비율 = \n",df_ratio2)
```
