## (1)Boston House Prices dataset 사용
   https://www.kaggle.com/에서 검색하여  데이터 특징 설명 및 파일
- CRIM : 도시 별 일인당 범죄율 
- ZN : 구역 별 주거용 토지 비율 25,000 평방 피트 이상 
- INDUS : 마을 당 비 소매업 지역의 비율 
- CHAS : 찰스 리버 더미 변수 (그렇지 않으면 관로 경계 인 경우 1, 그렇지 않은 경우 0) 
- NOx : 산화 질소 농도 (1000 만분의 1)
- RM : 주거 당 평균 객실 수 
- AGE : 1940 년 이전에 건설 된 소유주 점유 단위의 비율 
- DIS : 보스턴 고용 센터를위한 가중치가있는 거리 
- RAD : 방사형 고속도로 접근성 지수 
- TAX : 1 만달 러당 부당 이득 세율 
- PTRATIO : 마을 별 학생 - 교사 비율 
- B : 1000 (Bk - 0.63) ^ 2 여기서 Bk는 도시 별 흑인 비율
- LSTAT : 인구 감소 % 상태
- MEDV : 1000 달러짜리 주택 소유 주택의 중간 값


```python
from sklearn.datasets import load_boston

# Load dataset
#filename = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'
names = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
#dataset = pd.read_csv(filename, delim_whitespace=True, names=names)
boston = load_boston()
boston = pd.DataFrame(data=np.c_[boston['data'], boston['target']], columns=names )
```

```python
display(boston)
```

## (2)데이터의 형태(크기), 특징 이름, 상위 데이터 10개 확인

```python
print("데이터의 형태(크기) = \n")
display(boston.shape)

print("데이터의 index = \n")
display(boston.index)

print("데이터의 columns = \n")
display(boston.columns)

print("상위 데이터 10개 확인 = \n")
display(boston.head(10))

print("하위 데이터 10개 확인 = \n")
display(boston.tail(10))
```

## (3)모든 특징들의 분포 확인

```python
boston.hist(bins=5,figsize=(5,5))
```

## (4)모든 특징들 사이의 상관계수

```python
corr = boston.corr(method='pearson')
display(corr)
```

## (5)주택가격과 다른 특징들과의 관계 확인(MEDV와 다른 특징과의 상관계수)

```python
corrMEDV = corr['MEDV']
display(corrMEDV)
```

## (6) 주택가격과 다른 특징들에 대한 scatter plot

```python
for column in boston.columns:
    if column == "MEDV":
        continue
    boston.plot.scatter(x = column, y = "MEDV", title = column)
```
