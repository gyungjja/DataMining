# [실습 #9] 당뇨병 데이터 linear regression

## 1)diabetes 데이터 사용

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import sklearn.datasets as ds

#diabetes 데이터 불러오기
db = ds.load_diabetes()
```

## 2)데이터 설명(형태, 특징이름)

```python
#전체 출력
print("데이터의 형태 =\n",db.data.shape)
print("데이터의 특징이름=\n",db.feature_names)
print("데이터의 설명 = \n",db.DESCR)
```

## 3)데이터 분석

### (3-1)데이터프레임을 생성하여 타입, 요약통계, 상위 10개 확인

```python
print("=====데이터 프레임 생성=====")
dia = pd.DataFrame(db.data, columns=db.feature_names)
dia['diab'] = db.target
print("데이터의 타입 =\n")
display(dia.dtypes)
print("데이터의 요약 통계 = \n")
display(dia.describe())
print("상위 10개 확인 = \n")
display(dia.head(10))
```

### (3-2)개별 특징 분포확인 histogram

```python
dia.hist(bins=5, figsize=(9,9))
```

### (3-3)당뇨결과와 다른 특징들과의 선형 관계(scatter plot, correlation) 확인

```python
#correlation
corr = dia.corr(method='pearson')
display(corr)

corrdiabetes = corr['diab']
display(corrdiabetes)

#scatter plot
for i in dia.columns:
    dia.plot.scatter(x = i, y = 'diab', title= i)
```

### (3-4)당뇨결과를 제외한 특징들의 boxplot

```python
#diab열을 삭제한 데이터
dia1 = dia.drop(['diab'],axis=1)
display(dia1)

for i in dia1.columns:
    dia1.boxplot()
```

## 4)선형회귀

### (4-1)모든 특징 사용

```python
X = db.data
y = db.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print(X_train.shape)
print(X_test.shape)

lr = LinearRegression().fit(X_train, y_train)

w = lr.coef_
b = lr.intercept_
print("w (lr.coef_):", w)
print("b (lr.intercept_):",b)

print("훈련 세트 점수: {:.2f}".format(lr.score(X_train, y_train)))
print("테스트 세트 점수: {:2f}",format(lr.score(X_test, y_test)))
```

### (4-2)상관도가 높은 특징 4개 사용

```python
#correlation
corr = dia.corr(method='pearson')
display(corr)

corrdiabetes = corr['diab']
display(corrdiabetes)
#상관도가 높은 특징 4개 추출
sub = corrdiabetes[corrdiabetes.abs() > 0.4]
display(sub)

dia2 = dia.drop(['age','sex','s1','s2','s3','s6','diab'],axis=1)
display(dia2)

X_train, X_test, y_train, y_test = train_test_split(dia2, dia['diab'], random_state=0)
print(X_train.shape)
print(X_test.shape)

lr = LinearRegression().fit(X_train, y_train)

w = lr.coef_
b = lr.intercept_
print("w (lr.coef_):", w)
print("b (lr.intercept_):",b)

print("훈련 세트 점수: {:.2f}".format(lr.score(X_train, y_train)))
print("테스트 세트 점수: {: 2f}".format(lr.score(X_test, y_test)))
```

### (4-3)상관도가 높은 특징 3개 사용

```python
#correlation
corr = dia.corr(method='pearson')
display(corr)

corrdiabetes = corr['diab']
display(corrdiabetes)
#상관도가 높은 특징 4개 추출
sub = corrdiabetes[corrdiabetes.abs() > 0.44]
display(sub)

# bmi,bp,s5 만 추출
dia3 = dia2.drop(['s4'],axis=1)
display(dia3)

X_train, X_test, y_train, y_test = train_test_split(dia3, dia['diab'], random_state=0)
print(X_train.shape)
print(X_test.shape)

lr = LinearRegression().fit(X_train, y_train)

w = lr.coef_
b = lr.intercept_
print("w (lr.coef_):", w)
print("b (lr.intercept_):",b)

print("훈련 세트 점수: {:.2f}".format(lr.score(X_train, y_train)))
print("테스트 세트 점수: {: 2f}".format(lr.score(X_test, y_test)))
```

### (4-4)상관도 가장 높은 특징과의 scatter plot과 선형회귀직선

```python
#correlation
corr = dia.corr(method='pearson')
display(corr)

corrdiabetes = corr['diab']
display(corrdiabetes)

#선형회귀직선
X_train,X_test,y_train,y_test = train_test_split(dia2[['bmi']],db.target,random_state=0)

lr = LinearRegression().fit(X_train,y_train)
dia.plot.scatter(x = 'bmi', y = 'diab', marker= 'o' )
line = np.linspace(-0.1,0.2,100).reshape(-1,1)
plt.plot(line, lr.predict(line))
```
